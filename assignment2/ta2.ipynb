```python
import pandas as pd
import sqlite3 as sq
```

```python
conn = sq.connect('ta2.db')
```

```python
df = pd.read_csv('data/tips.csv')
df
```

```python
df.to_sql('tips', conn, if_exists='replace', index=False)
```

```python
cur = conn.cursor()
```

```python
for row in cur.execute('SELECT * FROM tips'):
    print(row)
```

```python
#Find the maximum and minimum total bill amounts
cur.execute('SELECT MAX(total_bill) FROM tips')
print("Max Value: ")
print(cur.fetchone()[0])

cur.execute('SELECT MIN(total_bill) FROM tips')
print("Min Value: ")
print(cur.fetchone()[0])
```

```python
#Retrieve the total bill and tip for parties of size 4 or more, where the tip percentage is greater than 15%
query = ''' SELECT total_bill, tip
FROM tips
WHERE size >= 4 AND (tip/total_bill)*100 > 15
'''

for row in cur.execute(query):
    print(row)
```

```python
#Find the average tip percentage for each combination of day, time, and smoker status

query = ''' SELECT day, time, smoker, ROUND(AVG((tip/total_bill)*100),2) AS tip_percent_avg
FROM tips
GROUP BY day, time, smoker
'''

for row in cur.execute(query):
    print(row)
```

```python
#Find the maximum and minimum tip percentage for each day and time combination, along with the corresponding total bill and tip amount

query = ''' SELECT total_bill, tip, day, time, smoker, ROUND(MAX((tip/total_bill)*100),2) AS tip_percent_max
FROM tips
GROUP BY day, time, smoker
'''

print("Maximums:")
for row in cur.execute(query):
    print(row)

query = ''' SELECT total_bill, tip, day, time, smoker, ROUND(MIN((tip/total_bill)*100),2) AS tip_percent_min
FROM tips
GROUP BY day, time, smoker
'''

print("Minimums:")
for row in cur.execute(query):
    print(row)
```

```python
#Find the average tip percentage for each combination of day, time, and smoker status, but only include combinations with more than 5 records
query = ''' SELECT day, time, smoker, ROUND(AVG((tip/total_bill)*100),2) AS tip_percent_avg
FROM tips
GROUP BY day, time, smoker
HAVING COUNT(*) > 5
'''

for row in cur.execute(query):
    print(row)
```
