- **systolic (mmHg)**: Represents the systolic blood pressure, which is the pressure in the arteries when the heart beats. Elevated systolic pressure can indicate risk for cardiovascular disease.

- **functional_changes_index (FCI)**: A computed index reflecting physiological adaptations to various external and internal stressors, such as illness, physical activity, or emotional stress.

- **circulatory_efficiency (CE)**: A derived index that assesses the efficiency of the cardiovascular system in delivering blood and oxygen to the body, providing an overview of cardiovascular health and performance.

- **kerdo_vegetation_index**: A metric related to the balance of the autonomic nervous system, particularly the sympathetic and parasympathetic systems. It gives insights into how the body manages stress and recovery.

- **robinson_index**: A calculated index that reflects cardiovascular performance by combining heart rate and blood pressure measurements. It is often used to estimate overall cardiac workload.

### **Handling Outliers**

Given the presence of substantial outliers, especially in circulatory_efficiency, it’s important to cap the extreme values to prevent them from distorting the results. We will use Winsorization, which caps outliers at the 5th and 95th percentiles.

````python
# Apply Winsorization to circulatory_efficiency to cap extreme outliers
data['circulatory_efficiency'] = winsorize(data['circulatory_efficiency'], limits=[0.05, 0.05])

# Visualize the updated box plot
plt.figure(figsize=(14, 8))
sns.boxplot(data=data[['diastolic', 'systolic', 'functional_changes_index', 'circulatory_efficiency', 'kerdo_vegetation_index', 'robinson_index']])
plt.title("Box Plot After Winsorizing Circulatory Efficiency")
plt.show()
````

### **Takeaways: Handling Outliers Using Winsorization**

To address these outliers, we applied **Winsorization**, a method that caps extreme values at the 5th and 95th percentiles. This ensures that the outliers are handled without removing critical data points, which might still carry valuable information.

After Winsorizing, we re-examined the distribution of the key variables to confirm that the outliers have been effectively managed, particularly in `circulatory_efficiency`.

### **a. Standardize the Data and find the Two Principal Components**

We will first standardize the data (as PCA is sensitive to the scale of the variables) and then apply PCA to reduce the dataset to 2 principal components.

````python
# Select relevant numerical columns for PCA
features = ['diastolic', 'systolic', 'functional_changes_index', 'circulatory_efficiency', 
            'kerdo_vegetation_index', 'robinson_index', 'pulse_pressure']

# Handling missing values by dropping rows with missing values (if any left, just to be safe :)!)
data_pca_input = data[features].dropna() 

# Standardize the features before applying PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_pca_input)

# Applying PCA without specifying components first to examine explained variance for all components
pca = PCA()
principal_components = pca.fit_transform(scaled_data)

# Creating a DataFrame with the explained variance for each principal component
explained_variance = pca.explained_variance_ratio_

# Plotting the explained variance and cumulative explained variance
plt.figure(figsize=(10, 6))

# Plot explained variance for each component
plt.subplot(1, 2, 1)
plt.bar(range(1, len(explained_variance) + 1), explained_variance, color='blue')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Principal Components')
plt.ylabel('Variance Ratio')

# Plot cumulative explained variance
plt.subplot(1, 2, 2)
cumulative_variance = explained_variance.cumsum()
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='blue')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance Ratio')

plt.tight_layout()
plt.show()

# Based on the cumulative variance, let's reduce to 3 components for a balance
pca_3comp = PCA(n_components=3)
principal_components_3comp = pca_3comp.fit_transform(scaled_data)

# Create a DataFrame for the principal components
pca_df_3comp = pd.DataFrame(data=principal_components_3comp, columns=['PC1', 'PC2', 'PC3'])

# Combine PCA results with the original data for analysis
data_pca_3comp = pd.concat([data_pca_input.reset_index(drop=True), pca_df_3comp], axis=1)

# Verify the explained variance ratio for 3 components
explained_variance_3comp = pca_3comp.explained_variance_ratio_
print(f"Explained Variance by PC1, PC2, and PC3: {explained_variance_3comp}")

# Combine PCA results with the original metadata (user_code and measurement_datetime)
data_pca_3comp = pd.concat([data[['user_code', 'measurement_datetime']].reset_index(drop=True), 
                             pca_df_3comp], 
                             axis=1)

# Display the reduced dataset with metadata
data_pca_3comp.head()
````

This allows us to retain the most important patterns in the data while reducing dimensionality, which simplifies further analysis or modeling.

- **Uncorrelated Components**: PCA transforms the original correlated variables into **uncorrelated principal components**, which makes it easier to work with the data, especially in machine learning models.
- **Improving Interpretability**: By reducing the number of variables, the dataset becomes easier to visualize, interpret, and process while preserving key insights.
